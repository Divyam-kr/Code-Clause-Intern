```python
#loading dataset 

df = pd.read_csv("Market_Basket_Optimisation.csv" , names=[i for i in range(0,20)])
```

```python
#The dataset shows the market basket of different customers while they bought items. Based on these combinations only 
#we will work out further

df
```

```python
#Replacing all null values with 0 to simplify our task
df.fillna(0,inplace=True)
```

```python
df
```

```python

#making a list where i will include all purchaces in each index of the list
purchases = []

#adding items of each purchase in the list as each index as a unique purchase
for i in range (0,len(df)):
    purchases.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j]) != '0'])
```

```python
#Viewing items purchased at 2nd number
purchases[1]
```

```python
purchases[34]
```

```python
#now importing and applying model 
from apyori import apriori

#Setting the hyperparameters 
model_0_5 = apriori(purchases, min_support=0.005, min_confidance=0.2, min_lift=3, min_length=2)

#min support of 0.5% with confidence 20 % and min length 2

model_0_2 = apriori(purchases, min_support=0.002, min_confidance=0.2, min_lift=3, min_length=2)
#finalising min support of 0.2% with confidence 20 % and min length 2 
#(Since it is found to be best on the basis of optimum number of association samples)
```

```python
model_0_5
```

```python
Result_0_5 = list(model_0_5)
```

```python
Result_0_5
```

```python
df_result_0_5 = pd.DataFrame(Result_0_5)
```

```python
#Showing the sample results based on min support 0.5%
df_result_0_5
```

```python
#Final result with o.2% min support
Result = list(model_0_2)
```

```python
df_final = pd.DataFrame(Result)
```

```python
df_final
```

```python
support = df_final.support
```

```python
#seperating the elements of ordered stats in correct order
lhs = []
rhs = []
Confidence = []
Lift = []

for i in range(len(df_final)):
    single_list = df_final['ordered_statistics'][i][0]
    lhs.append(list(single_list[0]))
    rhs.append(list(single_list[1]))
    Confidence.append(single_list[2])
    Lift.append(single_list[3])
```

```python
#Presenting then as a new Data Frame
df_final_result = pd.DataFrame(lhs) 
```

```python
df_final_result["rhs"] = rhs
```

```python
df_final_result["Support"] = support
```

```python
df_final_result["Confidence"] = Confidence
```

```python
df_final_result["Lift"] = Lift
```

```python
df_final_result
```

```python
#Rearranging the columns
df_final_result.columns = ['lhs',1,'rhs','Support','Confidence','Lift']
```

```python
df_final_result
```

```python
#Dropping the meaningless column
df_final_result.drop(columns=[1],inplace=True)
```
